```python
df_gsi = pd.read_json("gsi/8678700601_gsi.jsonl", lines=True, encoding="cp1251")
```

```python
df_gsi["hero"].apply(lambda x: x.get("name") if isinstance(x, dict) else None).unique()
```

```python
df_gsi["event_types"] = df_gsi["events"].apply(lambda x: [i.get("data") for i in x if isinstance(i, dict)] if len(x) > 0 else []).apply(lambda x: [ast.literal_eval(i).get("type") for i in x if isinstance(i, str)] if len(x) > 0 else []).apply(lambda x: set(x))
```

```python
mask = df_gsi["event_types"].apply(lambda s: "CHAT_MESSAGE_ITEM_PURCHASE" in s if isinstance(s, set) else False)
```
